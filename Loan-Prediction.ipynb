```python
X = df.drop('not.fully.paid',axis=1)
y = df['not.fully.paid']
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components=2)
X_pca = pca.fit_transform(df.drop('not.fully.paid', axis=1))

plt.figure(figsize=(10,6))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=df['not.fully.paid'], palette='Set1', alpha=0.6)
plt.title('PCA: 2D View of Data Colored by Default Status')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend(title='Not Fully Paid')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
from sklearn.linear_model import LogisticRegression
regression = LogisticRegression()
```

```python
regression.fit(X_train_scaled,y_train)
```

```python
y_pred = regression.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test,y_pred)
print("Accuracy :",accuracy)
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
```

```python
cm = confusion_matrix(y_test,y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

```python
#Trying to imporve results using SMOTE
```

```python
from imblearn.over_sampling import SMOTE
smote =SMOTE(random_state=42)
```

```python
X_train_resampled, y_train_resampled = smote.fit_resample(X_train,y_train)
```

```python
X_train_scaled_resample = scaler.fit_transform(X_train_resampled)
X_test_scaled_resample = scaler.transform(X_test)
```

```python
regression_resampled = LogisticRegression()
```

```python
regression_resampled.fit(X_train_scaled_resample,y_train_resampled)
```

```python
y_pred_resampled = regression_resampled.predict(X_test_scaled_resample)
```

```python
new_score = accuracy_score(y_test,y_pred_resampled)
print(new_score)
```

```python
cm = confusion_matrix(y_test,y_pred_resampled)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

```python
y_probs = regression_resampled.predict_proba(X_test_scaled_resample)[:,1]
```

```python
from sklearn.metrics import classification_report
threshold = 0.35
y_pred_thres = (y_probs >= threshold).astype(int)

print(confusion_matrix(y_test,y_pred_thres))
print(classification_report(y_test, y_pred_thres))
```

```python
print(accuracy_score(y_test,y_pred_thres))
```

```python
from sklearn.metrics import precision_recall_curve, auc
import matplotlib.pyplot as plt

precision, recall, thresholds = precision_recall_curve(y_test, y_probs)
plt.plot(recall, precision)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

fpr, tpr, thresholds = roc_curve(y_test, y_probs)
auc_score = roc_auc_score(y_test, y_probs)

plt.plot(fpr, tpr, label=f"AUC = {auc_score:.2f}")
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate (Recall)')
plt.title('ROC Curve')
plt.legend()
plt.show()
```
